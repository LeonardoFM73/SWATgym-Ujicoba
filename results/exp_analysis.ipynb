#### Evaluation

```python
# Performance of baseline methods on one full growing season.

random_rwds = random_rwds_seed0+random_rwds_seed1+random_rwds_seed2+random_rwds_seed3+random_rwds_seed4
standard_rwds = standard_rwds_seed0+standard_rwds_seed1+standard_rwds_seed2+standard_rwds_seed3+standard_rwds_seed4
reactive_rwds = reactive_rwds_seed0+reactive_rwds_seed1+reactive_rwds_seed2+reactive_rwds_seed3+reactive_rwds_seed4
ddpg_rwds = ddpg_rwds_seed0+ddpg_rwds_seed1+ddpg_rwds_seed2+ddpg_rwds_seed3+ddpg_rwds_seed4
td3_rwds = td3_rwds_seed0+td3_rwds_seed1+td3_rwds_seed2+td3_rwds_seed3+td3_rwds_seed4

plt.plot(random_rwds/5.0, label='random')
plt.plot(standard_rwds/5.0, label='standard')
plt.plot(ddpg_rwds/5.0, label='ddpg')
plt.plot(td3_rwds/5.0, label='td3')
plt.plot(reactive_rwds/5.0, label='reactive')

plt.xlabel('timestep (day)')
plt.ylabel('avg reward')
plt.legend()
```

```python
#  Learning curves of the RL algorithms on SWATGym

# DDPG
d_means = np.mean(d_arr, axis=0)
d_stds = np.std(d_arr, axis=0, dtype=np.float64)
x = np.arange(0, len(d_means))
plt.fill_between(x, d_means - d_stds/2.0, d_means + d_stds/2.0, alpha=0.1)
plt.plot(x, d_means, label='ddpg')

# TD3 
t_means = np.mean(t_arr, axis=0)
t_stds = np.std(t_arr, axis=0, dtype=np.float64)
x = np.arange(0, len(t_means))
plt.fill_between(x, t_means - t_stds/2.0, t_means + t_stds/2.0, alpha=0.1)
plt.plot(x, t_means, label='td3')

plt.xlabel('Episodes')
plt.ylabel('Average Return')
plt.legend()
```
